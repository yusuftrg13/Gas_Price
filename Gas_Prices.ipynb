```python
gas=pd.read_csv('Desktop/gas_prices.csv')

plt.title('DÜNYA GAZ FİYATLARINDA Kİ  DEĞİŞİMLER (USD)')


plt.plot(gas.Year, gas.USA, label='USA')
plt.plot(gas.Year, gas.Germany, label='Germany')  #Avrupa,Amerika,Asya kıtasının ekonomik gücünü çeken ülkelerde yıllara göre gaz fiyatları
plt.plot(gas.Year, gas.Japan, label='Japan')



print(gas.Year[::3])

plt.xticks(gas.Year[::3])

plt.xlabel('Yıl')

plt.ylabel('Fiyat(USD)')

plt.legend()


plt.grid(True)

plt.tight_layout()
```
